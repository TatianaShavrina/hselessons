### Time for some analytics

As it seems clear now, natasha shows different quality depending on text genre, normativeness and, more important, "russianness" of the name. This makes it a good tools for analyzing regional news and literature, but not the best tool for analyzing social media and world news, ans well as wikipedia and some text concerning foreign affairs.

For the "very Russian data" there are also two important tips:
1) natasha does not support surnames which have homonymy with adjectives
2) natasha does not understand cases, when a family is called by surname in plural
these two are rather typical situations, which makes it necessary to change the source code yourself if wanting to detect Russian names properly.

```python
#первая колонка - само слово, второе слово - количество букв, третья - явля ли палиндромом (yes\no)
stroka = input("Введите слово: ").lower()
output=open(r'C:\Users\USER\Downloads\vvod.csv', 'w', encoding="utf-8")
slovo = "хватит"
while stroka != slovo:
    output.write(stroka + ","+str(len(stroka))+",")
    if stroka ==stroka[::-1]:
        output.write("yes"+ "\n")
    else:
        output.write("no"+ "\n")
    stroka = input("Введите слово: ").lower()
else:
    print("Работа закончена!")
    output.close()
```

```python
# программа читает текст из файла и записывает такой же текстовый файл, но оставляет 
# на строчке максимум 10 слов, остальное переносит
f=open(r'C:\Users\USER\Downloads\lenta_test.txt', 'r', encoding="utf-8")
output=open(r'C:\Users\USER\Downloads\lenta_test_10words.txt', 'w', encoding="utf-8")
k = 10
for i in f:
    lst = i.split()
    if len(lst)<=k:
        output.write(i)
    else:
        koef = [i for i in range(0,(len(lst)//k)+1)]
        for j in range(len(koef)):
            a = lst[j*k:(j+1)*k:]
            stroka = " ".join(a)
            output.write(stroka+ "\n")
            #print(a)    

output.close()
```

```python
#программа читает слова от пользователя, пока не получит слово "хватит", 
#и затем определяет, является ли слово русской глагольной словоформой
stroka = input("Введите слово: ").lower()
output=open(r'C:\Users\USER\Downloads\glagoly.csv', 'w', encoding="utf-8")
slovo = "хватит"
suffix_list = ["ил","ила","или","ило","ал","али","ала","ало"]
okonch = ["щий", "щего", "щему", "щим", "щем", "щие", "щих", "щими"]
while stroka != slovo:
    for i in range(len(okonch)):
        if stroka.endswith(okonch[i]) and stroka.isalpha():
            output.write(stroka+ "," +"активное причастие"+"\n")
    
        if stroka.endswith(suffix_list[i]) and stroka.isalpha():
            output.write(stroka+ "," +"глагол прошед вр"+"\n")
        else:
            output.write(stroka+ "," +"это не глагол"+"\n")
    stroka = input("Введите слово: ").lower()
else:
    print("Работа закончена! Слово 'хватит' с вероятностью 1 является глаголом!")
    output.close()
```

```python
import os
#listd_hy=os.listdir(r'C:\Users\USER\Documents\НКРЯ\новости\article_hy')
listd_ru=os.listdir(r'C:\Users\USER\Documents\НКРЯ\новости\article_ru')
output=open(r'C:\Users\USER\Documents\НКРЯ\новости\hist.csv', 'w', encoding="utf-8")
lst = [] #список уникальных слов для каждого файла
lst_files=[]
dic = {} #общий словарь
dic_files = {} #словарь по количеству файлов, содержащих слово
for i in listd_ru:
    filename = r'C:\Users\USER\Documents\НКРЯ\новости\article_ru' + "\\" + i
    f = open(filename, 'r', encoding="utf-8" )
    lst = []
    for line in f:
        a = line.split() #сплиттим по пробелам
        for word in a:
            if word.isalpha():
                word = word.lower()
                #word = word.replace('<',"")
                #word = word.replace('>',"")
                #word = word.replace('/',"")
                #word = word.replace('\'',"")
                #word = word.replace(r'"',"")
                #word = word.strip("&laquo;")
                #word = word.strip("!--")
                #word = word.strip("(&laquo;")
                if word in dic: #проверка в качестве ключа
                    dic[word] += 1
                else:
                    dic[word] = 1
                if word not in lst:
                    lst.append(word)
    lst_files.append(lst)
        
for i in lst_files:
    for j in i:
        if j in dic_files: #проверка в качестве ключа
            dic_files[j] += 1
        else:
            dic_files[j] = 1
        
for word in sorted(dic):
    output.write(word +"," + str(dic[word])+"," +str(dic_files[word]) +"\n")
    print(word, dic[word])
output.close()
```

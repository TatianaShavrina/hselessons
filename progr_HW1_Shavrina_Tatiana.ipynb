(Помним, что таким образом мы получаем все типы смешения кириллицы и латиницы - а редактировать нам нужно отдельно, контекстно зависимо)

Видно, что ошибки вызваны, в основном, неправильным определением "i" как латиницы при OCR, а также квадратными скобками внутри слова. Но есть и более интересные ошибки, например, кириллическое "Х" в "XXI" веке, символы с ударениями и латинские "а" вместо кириллических. 

Анализ ошибок также показывает, что в кириллицу необходимо будет добавить ять, i кириллическое.

Правила замены, таким образом, будут такие:

**Контекстно независимые:**
1. уберем квадратные скобки внутри слова

**Конеткстно зависимые:**
1. примеры с русскими окончаниями после латинской буквы и апострофа пропускаем, они "легитимны"
+ в соседстве кириллических букв i латинское заменим на кириллическое - обратную замену проведем в случае кириллического i в окружении латиницы
+ в соседстве кириллических букв буквы лат.алфавита, случайно попавшие в русское слово - на их русское соответствие (заметим, что таких замен потенциально небольшое количество: a - а, o - о, c - с, и т.д.). Аналогично замену проведем в случае кириллических букв из списка в окружении латиницы


При таком наборе правил останется некоторое кол-во исключений, когда сочетание кириллицы и латиницы вполне нормативное: это случаи с дефисом и апострофом: St.-Jérôme’oмъ, Cis-moll-ная и т.д. Мы также не трогаем случаи с акутом и грависом во всех словах.

```python
#создадим файлы для записи исправленного варианта (копия старого) и документации исправлений

corrected = r'C:\Users\USER\Documents\hse\progr' + "\\" + link[-6:-1].replace("/","") + "-corr" + ".html"
#output_copy = shutil.copy(out2, corrected)
output_corrected = open(corrected, 'w', encoding="utf-8")
# в документацию ошибок будем писать порядковый номер слова в тексте, исходный вариант и на что заменили - через табуляцию
errors2 = r'C:\Users\USER\Documents\hse\progr' + "\\" + link[-6:-1].replace("/","") + "-errors" + ".txt"
errors_corrected = open(errors2, 'w', encoding="utf-8")
```

```python
def skobki(i, word):
    stroka = str(i) + "\t" + word + "\t"
    word = word.replace(r'[',"")
    word = word.replace(r']',"")
    stroka = stroka + word + "\n"
    errors_corrected.write(stroka)
    return word
```

```python
# контекстно независимая замена
for i in range(len(wordlist)):
    if  "<" not in wordlist[i] and ">" not in wordlist[i] and  ";" not in wordlist[i]:
        if re.search(r'[[\w+]]', wordlist[i]):
            wordlist[i] = skobki(i, wordlist[i])
```

```python
def alphabet(i, word):
    
    result = word
    stroka = str(i) + "\t" + word + "\t"
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+i', word): # i кириллическое
        result = re.sub(r'i', r"і", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+I', word):
        result = re.sub(r'I', r"І", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+a', word): # a 
        result = re.sub(r'a', r"а", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+A', word):
        result = re.sub(r'A', r"А", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+e', word): #e 
        result = re.sub(r'e', r"е", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+E', word):
        result = re.sub(r'E', r"Е", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+o', word): # o
        result = re.sub(r'o', r"о", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+O', word):
        result = re.sub(r'O', r"О", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+u', word): # u 
        result = re.sub(r'u', r"и", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+U', word):
        result = re.sub(r'U', r"И", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+y', word): # y
        result = re.sub(r'y', r"у", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+Y', word):
        result = re.sub(r'Y', r"У", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+c', word): # с
        result = re.sub(r'c', r"с", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+C', word):
        result = re.sub(r'C', r"С", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+m', word): # m
        result = re.sub(r'm', r"т", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+M', word): # M
        result = re.sub(r'M', r"М", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+x', word): # x
        result = re.sub(r'x', r"х", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+X', word):
        result = re.sub(r'X', r"Х", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+p', word): # p
        result = re.sub(r'p', r"р", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+P', word):
        result = re.sub(r'P', r"Р", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+b', word): # bь
        result = re.sub(r'b', r"ь", word)
    if re.search(r'[а-яА-ЯёЁѣѢіІ]+r', word): # rг
        result = re.sub(r'r', r"г", word)
    
    # а теперь убираем путаницу с латиницей
    # по примерам, в первую очередь чаще всего кириллическое "х" попадает на 1ое место в "XXI" и т.д., а также
    # есть одна из самых частотных ошибок - "за" c латинской а. Поэтому для случаев, где контекста слева - 1 буква,
    # будет доп.условие на контекст справа
    
    if re.search(r'Х[a-zA-Z]+[a-zA-Z]', word):
        result = re.sub(r'Х', r"X", word)
    if re.search(r'[a-zA-Z]+а[a-zA-Z]+', word): # a 
        result = re.sub(r'а', r"a", word)
    if re.search(r'[a-zA-Z]+A[a-zA-Z]', word):
        result = re.sub(r'А', r"A", word)
    if re.search(r'[a-zA-Z]+е[a-zA-Z]', word): #e 
        result = re.sub(r'е', r"e", word)
    if re.search(r'[a-zA-Z]+Е[a-zA-Z]', word):
        result = re.sub(r'Е', r"E", word)
    if re.search(r'[a-zA-Z]+о[a-zA-Z]', word): # o
        result = re.sub(r'о', r"o", word)
    if re.search(r'[a-zA-Z]+О[a-zA-Z]', word):
        result = re.sub(r'О', r"O", word)
    if re.search(r'[a-zA-Z]+и[a-zA-Z]', word): # u 
        result = re.sub(r'и', r"u", word)
    if re.search(r'[a-zA-Z]+у[a-zA-Z]', word): # y
        result = re.sub(r'у', r"y", word)
    if re.search(r'[a-zA-Z]+У[a-zA-Z]', word):
        result = re.sub(r'У', r"Y", word)
    if re.search(r'[a-zA-Z]+с[a-zA-Z]', word): # с
        result = re.sub(r'с', r"c", word)
    if re.search(r'[a-zA-Z]+С[a-zA-Z]', word):
        result = re.sub(r'С', r"C", word)
    
    word = result
    stroka = stroka + word + "\n"
    errors_corrected.write(stroka)
    print(stroka)
    return word
```

```python
# контекстно зависимая замена 
                     
for i in range(len(wordlist)):
    if  "<" not in wordlist[i] and ">" not in wordlist[i] and  ";" not in wordlist[i]:
        if re.search(r'[a-zA-Z]+[а-яА-ЯёЁѣѢіІ]', wordlist[i]) or re.search(r'[а-яА-ЯёЁѣѢіІ]+[a-zA-Z]', wordlist[i]):
            if bool(re.search(r"[a-zA-Z]['|’][а-яА-Яёѣі]", wordlist[i]))==False: # тут примеры типа "St.-Jérôm’омъ"
                wordlist[i] = alphabet(i, wordlist[i])
```

```python
# записываем в файл output_corrected отредактированный текст

text3 = " ".join(wordlist)
for line in text3:
    output_corrected.write(line)
```
